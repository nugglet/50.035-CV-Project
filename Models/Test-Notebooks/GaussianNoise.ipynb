#Gaussian Noise 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def addGaussianNoise(img, sigma):
    row,col,ch= img.shape
    mean = 0
    var = 0.1
    gauss = np.random.normal(mean,sigma,(row,col,ch))
    gauss = gauss.reshape(row,col,ch)
    noisy = img + gauss
    noisy /= 255
    return noisy

image = cv2.imread('blackandwhite1.jpeg')
plt.imshow(image),
plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.show()
image_noise = addGaussianNoise(image,20)
plt.imshow(image_noise),
plt.title('Noisy, sigma = 20'),plt.xticks([]), plt.yticks([])
plt.show()
image_noise1 = addGaussianNoise(image,50)
plt.imshow(image_noise1),
plt.title('Noisy, sigma = 50'),plt.xticks([]), plt.yticks([])
plt.show()
```
